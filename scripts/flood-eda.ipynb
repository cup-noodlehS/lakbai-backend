```python
from shapely.geometry import Point

point = Point(123.94285641475065,10.327898897323992)

is_included = False

for idx, row in shapefile.iterrows():
    if row['geometry'].contains(point):
        print(f"Point {point} is inside geometry at index {idx}")
        is_included = True
        break

if not is_included:
    print("Point does not belong to any of the indices")
```
